```python
def generate_dataset(Name):
    face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    
    def face_cropped(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
        
        if len(faces) == 0:
            return None
        for (x, y, w, h) in faces:
            cropped_face = img[y:y+h, x:x+w]
        return cropped_face
    
    cap = cv2.VideoCapture(1)
    if not cap.isOpened():
        print("Error: Could not open video capture")
        return
    
    img_id = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            break
        
        cropped_face = face_cropped(frame)
        if cropped_face is not None:
            img_id += 1
            face = cv2.resize(cropped_face, (200, 200))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
            file_name_path = "C:/Vaibhav Data/Proffesional/Engineering/NHCE/Sem 6/Mini Project - III/Face_recog/data/"+ Name + str(img_id) + ".jpg"
            cv2.imwrite(file_name_path, face)
            cv2.putText(face, str(img_id), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
            
            cv2.imshow("Cropped_Face", face)
            if cv2.waitKey(1) == 13 or img_id == 250:
                break
        else:
            cv2.imshow("Camera Feed", frame)
            if cv2.waitKey(1) == 13:
                break
    
    cap.release()
    cv2.destroyAllWindows()
    print("Collecting samples is completed.")

# generate_dataset("Vaibhav")
# time.sleep(10)
# generate_dataset("Sruthi")
# time.sleep(10)
generate_dataset("Kamran")
time.sleep(10)
generate_dataset("Karthik")
```

```python
import os

def find_images(directory, extensions=None):
    """
    Finds all images in a given directory with specified extensions.
    
    Args:
        directory (str): Path to the directory to search for images.
        extensions (list): List of valid image file extensions (default: common formats).
    
    Returns:
        list: List of paths to image files.
    """
    if extensions is None:
        # Default to common image extensions
        extensions = ['.jpg', '.jpeg', '.png']
    
    # Normalize extensions to lowercase
    extensions = [ext.lower() for ext in extensions]
    
    # List to hold image paths
    image_files = []
    
    # Walk through the directory
    for root, _, files in os.walk(directory):
        for file in files:
            if os.path.splitext(file)[1].lower() in extensions:
                image_files.append(os.path.join(root, file))
    
    return image_files

# Example usage
directory_path = "indian_faces"  # Replace with your directory path
images = find_images(directory_path)

print("Found Images:")
count = 0
for image in images:
    count += 1
    print(count, image)

```

```python
import cv2
import os

def extract_faces(image_path, output_dir, count,resize_dim=(200, 200)):
    """
    Extracts faces from an image, converts them to grayscale, resizes them, 
    and saves them as 50x50 images.
    
    Args:
        image_path (str): Path to the input image.
        output_dir (str): Directory to save the extracted face images.
        resize_dim (tuple): Dimensions to resize the face images (default: (50, 50)).
    """
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Unable to load image {image_path}")
        return
    
    # Convert to grayscale for face detection
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Load pre-trained Haar Cascade for face detection
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    if len(faces) == 0:
        print("No faces detected in the image.")
        return
    
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)
    
    # Process each detected face
    for i, (x, y, w, h) in enumerate(faces):
        # Extract the face region
        face = gray_image[y:y+h, x:x+w]
        
        # Resize to the desired dimensions
        resized_face = cv2.resize(face, resize_dim)
        
        # Save the resized grayscale face image
        output_path = os.path.join(output_dir, f"intruder_{count+i+1}.png")
        cv2.imwrite(output_path, resized_face)
        print(f"Saved face {i+1} to {output_path}")

directory_path = "indian_faces"  # Replace with your directory path
images = find_images(directory_path)

print("Found Images:")
for idx, image in enumerate(images):
    extract_faces(image, "data", count=idx+1)
```
